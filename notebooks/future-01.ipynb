```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_22 = pd.read_csv('../data/2022/level_3.csv')
df_22.head()
```

```python
df_22.info()
```

```python
not_english_data = pd.read_csv('../data/2023/not_english.csv')
english_data = pd.read_csv('../data/2023/english.csv')
not_english_data.info()
```

```python
not_english_prn = list(not_english_data['Provider name'].unique())
english_prn = list(english_data['Provider name'].unique())
df_22_prn = list(df_22['Provider'].unique())
```

```python
# for val in not_english_prn:
for val in english_prn:
  if val in df_22_prn:
    print(val)
```

```python
df_22.head()
```

```python
def get_region(provider):
  if provider in not_english_prn:
    return 'Not-English'
  elif provider in english_prn:
    return 'English'
  else:
    return 'Other'
```

```python
df_22['Region'] = df_22['Provider'].apply(get_region)
df_22.head()
```

```python
df_22['Region'].value_counts()
```

```python
english_22 = df_22[df_22['Region'] == 'English']
not_english_22 = df_22[df_22['Region'] == 'Not-English']
```

```python
sample = not_english_22[['Answered 1', 'Answered 2', 'Answered 3', 'Answered 4', 'Answered 5', 'Actual value']]
sample.head()
```

```python
sample['Answered 1'] = sample['Answered 1'].str.replace(r'%', '')
sample['Answered 2'] = sample['Answered 2'].str.replace(r'%', '')
sample['Answered 3'] = sample['Answered 3'].str.replace(r'%', '')
sample['Answered 4'] = sample['Answered 4'].str.replace(r'%', '')
sample['Answered 5'] = sample['Answered 5'].str.replace(r'%', '')
```

```python
sample['Answered 1'] = pd.to_numeric(sample['Answered 1'])
sample['Answered 2'] = pd.to_numeric(sample['Answered 2'])
sample['Answered 3'] = pd.to_numeric(sample['Answered 3'])
sample['Answered 4'] = pd.to_numeric(sample['Answered 4'])
sample['Answered 5'] = pd.to_numeric(sample['Answered 5'])
```

```python
sample.head()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

```python
sample.info()
```

```python
27837 / 27
```

```python
data = sample.values.reshape((1031, 27, 6))
X = data[:, :26, :5].reshape(1031, 26*5)
y = data[:, 26, 5]
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=102)
```

```python
print(X_test[1])
```

```python
print(y_test[1])
```

```python
model = Sequential()
model.add(Dense(128, input_dim=26*5, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.1))
model.add(Dense(32, activation='relu'))
# model.add(Dense(16, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))

loss = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
```

```python
predictions = model.predict(X_test)
predictions_train = model.predict(X_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae_train = mean_absolute_error(y_train, predictions_train)
mse_train = mean_squared_error(y_train, predictions_train)
r2 = r2_score(y_train, predictions_train)

# Print performance metrics
print(f"Mean Absolute Error (MAE): {mae_train}")
print(f"Mean Squared Error (MSE): {mse_train}")
print(f"R Score: {r2}")

# fig, (ax1, ax2) = plt.subplots(1, 2)


# Plot actual vs predicted values
plt.figure(figsize=(5, 4))
plt.scatter(y_train, predictions_train, alpha=0.5, label='Predicted vs Actual')
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], color='red', label='Ideal Fit')
plt.xlabel('Actual Values',fontsize=15)
plt.ylabel('Predicted Values',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.title('Train R-squared: {}'.format(round(r2, 2)), fontsize=22, weight='bold')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

# Print performance metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R Score: {r2}")

# Plot actual vs predicted values
plt.figure(figsize=(5, 4))
plt.scatter(y_test, predictions, alpha=0.5, label='Predicted vs Actual')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Ideal Fit')
plt.xlabel('Actual Values',fontsize=15)
plt.ylabel('Predicted Values',fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.title('Test R-squared: {}'.format(round(r2, 2)), fontsize=22, weight='bold')
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

# Plot residuals
residuals = y_test - predictions.flatten()
plt.figure(figsize=(8, 4))
plt.scatter(predictions, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Values')
plt.grid(True)
plt.show()
```
