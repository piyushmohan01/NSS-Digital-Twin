```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
english = pd.read_csv('../data/2023/english_predicted.csv')
not_english = pd.read_csv('../data/2023/not_english.csv')
english.head()
```

```python
data_23 = pd.read_csv('../data/2023/NSS3_2023.csv')
data_23.head()
```

```python
eng_providers = list(english['Provider'].unique())
```

```python
rows_list = []
for i in range(len(data_23)):
  if data_23['Provider name'].values[i] in eng_providers:
    rows_list.append(dict(data_23.iloc[i]))
```

```python
combined_df = pd.DataFrame(rows_list)
```

```python
combined_df.head()
```

```python
not_english.head()
```

```python
combined_df.rename(columns={'Positivity measure (%)': 'Positivity', 'Publication response rate (%)': 'Response Rate'}, inplace=True)
```

```python
combined_df[['Question Num', 'Question']] = combined_df['Question'].str.split(':', expand=True)
```

```python
not_english.drop(columns='index1', inplace=True)
```

```python
cols_to_keep = list(not_english.columns)
combined_df = combined_df[cols_to_keep]
combined_df.head()
```

```python
combined_df.shape
```

```python
combined_df = pd.concat([combined_df, not_english], axis=0)
combined_df.head()
```
