```python
level_3['Actual value'] = level_3['Actual value'].str.replace(r'%', '')
level_3['Actual value'] = pd.to_numeric(level_3['Actual value'])
level_3.head()
```

```python
# for each programme (subject code):
#   get avg response and avg sample size
#   avg resp / avg sample = comp rate
#   add subject code to dictionary as key
#   add actual value and comp rate as values
```

```python
def get_column_by_comprate(col, dic):
  col_list = list(level_3[col].unique())
  comp_rate_list = []
  actual_val_list = []

  for value in col_list:
    counter = 0
    sum_response, sum_sample, sum_actual = 0, 0, 0
    df = level_3[level_3[col] == value]
    for row in range(len(df)):
      sum_response += df.iloc[row]['Response']
      sum_sample += df.iloc[row]['Sample Size']
      sum_actual += df.iloc[row]['Actual value']
      counter += 1
    if counter == 0:
      print(value)
      break
    
    avg_actual = round(sum_actual/counter, 2)
    avg_completion_rate = round(sum_response / sum_sample, 2)
    comp_rate_list.append(avg_completion_rate)
    actual_val_list.append(avg_actual)
    dic[value] = [avg_completion_rate, avg_actual]
  return dic, comp_rate_list, actual_val_list
```

```python
comp_rates_by_subject = {}
comp_rates_by_subject, comprate_subject, actualval_subject = get_column_by_comprate('Subject Code', comp_rates_by_subject)
subject_by_completion = pd.DataFrame({'subject': list(comp_rates_by_subject.keys()),
                                   'comp-rate': comprate_subject,
                                   'actual-val': actualval_subject})
subject_by_completion.to_csv('../data/2022/subject-comp-rate-vs-actual-agree.csv')
```

```python
comp_rates_by_provider = {}
comp_rates_by_provider, comprate_provider, actualval_provider = get_column_by_comprate('Provider', comp_rates_by_provider)
provider_by_completion = pd.DataFrame({'provider': list(comp_rates_by_provider.keys()),
                                   'comp-rate': comprate_provider,
                                   'actual-val': actualval_provider})
provider_by_completion.to_csv('../data/2022/provider-comp-rate-vs-actual-agree.csv')
```

```python
plt.style.use('fivethirtyeight')
ax = sns.lmplot(x='comp-rate', y='actual-val', data=subject_by_completion, legend=False, height=4, aspect=2)
plt.title('Comparison by Subject (2022)', fontsize=20, weight='bold')
plt.xlabel("Completion-Rate",fontsize=15)
plt.ylabel("Positivity",fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.ylim(top=100)
plt.show()
```
